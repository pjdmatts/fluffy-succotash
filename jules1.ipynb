```julia
using Primes

# return the next prime after x

# define the function...
function next_prime(x)
    x += iseven(x) ? 1 : 2
    while !isprime(x)
        x += 2
    end
    return x
end
```

```julia
import Pkg; Pkg.add("Primes")
```

```julia
next_prime(10000000)
```

```julia
next_prime(3)
```

```julia
next_prime(100)
```

```julia
# for-loop factorial

function factorial_for(n)
    fact = 1
    for i in 1:n         # i takes each value in 1, 2, ..., n
        fact *= i        # equivalent to fact = fact * i
    end
    return fact
end
```

```julia
factorial_for(10)
```

```julia
# factorial via range product

function factorial_range(n)
    return prod(1:n)
end
```

```julia
factorial_range(10)
```

```julia
# recursive definition of factorial

function factorial_rec(n)
    if n < 2
        return 1
    else
        return n * factorial_rec(n - 1)
    end
end
```

```julia
factorial_rec(10)
```

```julia
# recursive factorial with ternary operator

function factorial_rec_ternary(n)
    n < 2 ? 1 : n * factorial_rec_ternary(n - 1)
end
```

```julia
factorial_rec_ternary(10)
```

```julia
factorial_rec_ternary(100)
```

```julia
factorial_rec_ternary(10)
```

```julia
factorial_rec_ternary(11)
```
